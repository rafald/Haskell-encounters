```haskell
import Control.Exception
import Control.Monad.Trans.Cont
import Control.Monad.Identity

:t sequence

--nest :: forall a r (m :: * -> *). [(a -> m r) -> m r] -> ([a] -> m r) -> m r
nest :: (Monad m) => [(a -> m r) -> m r] -> ([a] -> m r) -> m r
nest xs = runContT (sequence (map ContT xs))

-- Following would cause unification of a and m r
-- \ar -> ar $ ar 's'
m = nest [\ar -> ar 7, \ar -> ar 11, \ar -> ar 13]
:t m
res = m $ (\as -> sum as) -- now m r is Integer
:t res
print (res :: Identity Integer)
```
