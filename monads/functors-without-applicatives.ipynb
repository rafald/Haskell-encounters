# examples-of-functors-without-applicatives

```haskell
{-# LANGUAGE ExistentialQuantification ,FlexibleContexts ,DeriveFunctor ,StandaloneDeriving 
              #-}
-- Non type-variable argument in the constraint: Num (AFunctor t) 
-- (Use FlexibleContexts to permit this)
```

(https://stackoverflow.com/questions/22099425/examples-of-functors-without-applicatives)

```haskell
data AFunctor a = forall f . Functor f => AFunctor { unFunct :: f a } 
deriving instance Functor a => Functor (AFunctor )
```

```haskell
af = AFunctor [2,3,5,7,11]
:t af
xs = fmap (*3) (unFunct af)
:t xs
p = fmap (*3) AFunctor (2,3)
:t p
fmap show p
```
